```python
from groq import Groq

groq_api_key = "Your GROG API KEY"

client = Groq(api_key=groq_api_key)

# Die Frage des Bauleiters auf der Baustelle
frage = "Ich habe gerade den Beton gegossen. Was sind die nГ¤chsten Schritte laut Protokoll?"

# Der System-Prompt zwingt LLaMA, nur auf SGR-Daten zu antworten
prompt = f"""
Du bist der interne KI-Assistent von StadtgrГјn Ruhr.
Beantworte die Frage des Mitarbeiters ausschlieГџlich basierend auf dem bereitgestellten Dokument.
Antworte immer auf Deutsch.

Dokument:
{sgr_dokument}

Frage:
{frage}
"""

print("Sende Anfrage an den Groq SGR-Bot (LLaMA-3.1)...")

try:
    completion = client.chat.completions.create(
        model="llama-3.1-8b-instant",
        messages=[
            {
                "role": "user",
                "content": prompt
            }
        ],
        temperature=0.1, # Niedrige Temperature fГјr faktenbasierte Antworten
        max_tokens=500
    )

    print("\n=== SGR MITARBEITER-APP (GROQ KI-BOT ANTWORT) ===")
    print(completion.choices[0].message.content)

except Exception as e:
    print(f"\nEin Fehler ist aufgetreten: {e}")
```
